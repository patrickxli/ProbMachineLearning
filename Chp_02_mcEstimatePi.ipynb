# Estimate Pi by Monte-Carlo method

```python
import matplotlib.pyplot as plt
import numpy as np


def mcEstimatePi(n, is_plot=False):
    x = 2 * np.random.rand(n, 1) - 1
    y = 2 * np.random.rand(n, 1) - 1
    in_circle = x ** 2 + y ** 2 <= 1
    h = np.sum(in_circle)
    sigma_squared = np.sum((in_circle - np.pi) ** 2) / n
    if is_plot:
        ax = plt.gca()
        ax.set_aspect('equal')
        plt.plot(x, y, 'r.', alpha=0.5)
        plt.plot(x[in_circle], y[in_circle], 'b.', alpha=0.5)
        # ax.add_patch(plt.Circle((0, 0), radius=1, fill=False, linewidth=0.5, ls='-.'))
        plt.show()
    return {"estimate": 4 * h / n, "standard error": np.sqrt(sigma_squared / n)}
```

```python
# plot the points and the unit circle area
N = 10000
print(mcEstimatePi(N, True))
```

```python
n_power = np.arange(1, 6, 1)

N = np.power(10, n_power)
for i in range(N.shape[0]):
    estimate_result = mcEstimatePi(N[i])
    print("{:.0f} Monte-Carlo trails, estimate Pi = {:.4f},"
          "standard error = {:.4f}\n".format(N[i], estimate_result["estimate"], estimate_result["standard error"]))
```
